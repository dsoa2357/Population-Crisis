## 5-3 인구 소멸 위기 지역 계산하고 데이터 정리하기

* 인구 소멸 위기란?
    * 65세 이상 노인 인구와 20~39세 여성 인구를 비교하여 여성 인구가 노인 인구의 절반에 미달할 경우

```python
# 20~39세 값을 더해 준다.
population['20-39세'] = population['20 - 24세'] + population['25 - 29세'] + population['30 - 34세'] + population['35 - 39세']

# 65세 이상 값을 더해 준다.
population['65세이상'] = population['65 - 69세'] + population['70 - 74세'] + \
                        population['75 - 79세'] + population['80 - 84세'] + \
                        population['85 - 89세'] + population['90 - 94세'] + \
                        population['95 - 99세'] + population['100+'] 

population.head(10)
```

```python
# index는 '광역시도', '시도'로 하고 컬럼은 '구분', 안에 들어가는 값들은 '인구수', '20-39세', '65세이상'이 들어간다.
pop = pd.pivot_table(population, 
                    index = ['광역시도', '시도'],
                    columns = ['구분'],
                    values = ['인구수', '20-39세', '65세이상'])
pop
```

```python
# 인구 소멸 비율: 20~30대 여자 / (65세이상/2), 소멸비율은 파생변수  ※ 파생변수란? 원래의 변수가 아닌 새롭게 만들어 내는 변수
# 소멸위기지역 : 소멸비용 < 1.0

pop['소멸비율'] = pop['20-39세', '여자'] / (pop['65세이상', '합계'] / 2)
pop.head()
```

```python
# '소멸위기지역' 컬럼을 새로 만들고 '소멸비율' 컬럼의 값이 < 1이면 T/F가 뜨게끔 한다.
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop.head()
```

```python
# '소멸위기지역' 컬럼의 값이 True면 그에 맞는 '시도' 컬럼이 출력
# multiful columns에서 값 뽑기: df.index.get_level_values(1)을 이용하여 소멸위기지역을 뽑아낼 수 있다.

pop[pop['소멸위기지역']==True].index.get_level_values(1)
```

```python
# pivot_table에 의해 다단으로 구성된 인덱스를 다시 초기화

pop.reset_index(inplace=True)
pop.head()
```

```python
# '20~39세'.. 등 합쳐 있던 컬럼을 나누어 준다.

tmp_columns = [pop.columns.get_level_values(0)[n] + \
              pop.columns.get_level_values(1)[n]
              for n in range(0, len(pop.columns.get_level_values(0)))]

pop.columns = tmp_columns

pop.head()
```

```python
# 'pop' 데이터 프레임의 정보를 간략하게 본다.
pop.info()
```

## 5-5 지도 시각화를 위해 지역별 고유 ID 만들기

* Folium 모듈을 사용하여 그리고 경게선을 그려주는 json 파일을 사용

```python
# '시도' 컬럼에 있는 고유값을 본다. -> 모든 지역을 대표하는 고유한 ID를 만든다.
pop['시도'].unique()
```

```python
si_name = [None] * len(pop)

# 광역시가 아닌데 구를 가지고 있는 시와 그 행정구를 파이썬의 dict형으로 선언하여 따로 설정을 해 줌

tmp_gu_dict = {'수원':['장안구', '권선구', '팔달구', '영통구'],
               '성남':['수정구', '중원구', '분당구'], 
               '안양':['만안구', '동안구'], 
               '안산':['상록구', '단원구'], 
               '고양':['덕양구', '일산동구', '일산서구'], 
               '용인':['처인구', '기흥구', '수지구'], 
               '청주':['상당구', '서원구', '흥덕구', '청원구'], 
               '천안':['동남구', '서북구'], 
               '전주':['완산구', '덕진구'], 
               '포항':['남구', '북구'], '창원':['의창구', '성산구', '진해구', '마산합포구', '마산회원구'], 
               '부천':['오정구', '원미구', '소사구']}
```

```python
for n in pop.index:
    # '광역시도' 컬럼 위에서 3자리가 '광역시', '특별시', '자치시'가 아닌 경우에
    if pop['광역시도'][n][-3:] not in ['광역시', '특별시', '자치시']:
        # '시도' 컬럼에서 '고성'이거나 '광역시도' 컬럼에서 '강원도'일 경우
        if pop['시도'][n][:-1] == '고성' and pop['광역시도'][n] == '강원도':
            # 이름을 '고성(강원)'으로 해 준다.
            si_name[n] = '고성(강원)'
        # 또는 '시도' 컬럼에서 '고성'이거나 '광역시도' 컬럼에서 '경상남도'일 경우,
        elif pop['시도'][n][:-1]=='고성' and pop['광역시도'][n]=='경상남도':
            si_name[n] = '고성(경남)'
        # 그 외의 경우에는 그냥 둔다.
        else:
            si_name[n] = pop['시도'][n][:-1]
            
        # tmp_gu_dict에 들어있는 수만큼 for문을 돌려준다.
        for keys, values in tmp_gu_dict.items():
            # 만약 values가 '시도' 컬럼에 있다면
            if pop['시도'][n] in values:
                # '시도' 컬럼에 있는 글자 수가 2와 같다면
                if len(pop['시도'][n])==2:
                    # ' '(공백)과 '시도' 컬럼에 있는 값을 더한 값을 넣어준다.
                    si_name[n] = keys + ' ' + pop['시도'][n]
                # 또는 '시도' 컬럼에 '마산합포구', '마산회원구'가 있으면
                elif pop['시도'][n] in ['마산합포구', '마산회원구']:
                    # ' '(공백)과 '시도' 컬럼에 있는 값을 더한 값을 넣어준다.
                    si_name[n] = keys + ' ' + pop['시도'][n][2:-1]
                # 그 외에는
                else:
                    # ' '(공백)과 '시도' 컬럼에 있는 값을 더한 값을 넣어준다.
                    si_name[n] = keys + ' ' + pop['시도'][n][:-1]
    # 또는 '광역시도' 컬럼에 '세종특별자치시'가 있으면
    elif pop['광역시도'][n] == '세종특별자치시':
        # '세종'으로 변경한다.
        si_name[n] = '세종'
        
    
    # 그 외에
    else:
        if len(pop['시도'][n])==2:
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n]
        else:
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n][:-1]
```

```python
# 지도에 표시하기 위해서 고유 'ID'를 만들어 준다.
si_name
```

```python
pop['ID'] = si_name
pop.head(3)
```

```python
# 필요없는 데이터 컬럼을 지운다.
# '20-39세 남자', '65세이상남자', '65세이상여자' 컬럼을 삭제한다.

del pop['20-39세남자']
del pop['65세이상남자']
del pop['65세이상여자']

pop.head()
```

```python
# stack(): x, y의 좌표를 얻기 위한 함수
draw_korea_raw_stacked = pd.DataFrame(draw_korea_raw.stack())

# index를 다시 설정해준다.
draw_korea_raw_stacked.reset_index(inplace=True)

# 컬럼의 값을 설정해 준다.
draw_korea_raw_stacked.rename(columns={'level_0':'y', 'level_1':'x', 0:'ID'},
                             inplace=True)
draw_korea_raw_stacked.head(15)
```
